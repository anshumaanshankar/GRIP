```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
wcss=[]
x=df.iloc[:,:].values
for i in range (1,11):
  from sklearn.cluster import KMeans
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title("The Elbow Method")
plt.show()
```

The major deviation is seen at 3 clusters, so we will assume 3 clusters for the given dataset

```python
kmeans=KMeans(n_clusters= 3)
y_kmeans=kmeans.fit_predict(x)
(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1])
```

VISUALISING THE CLUSTERS ALONG WITH THEIR CENTROIDS

```python
plt.scatter(x[y_kmeans ==0,0], x[y_kmeans == 0,1], color = 'cyan', label ='Cluster 0')
plt.scatter(x[y_kmeans ==1,0], x[y_kmeans == 1,1], color = 'magenta' , label ='Cluster 1')
plt.scatter(x[y_kmeans ==2,0], x[y_kmeans == 2,1], color = 'yellow', label ='Cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = 'black', label ='Centroid')
plt.title('Clustered datapoints')
plt.legend()
plt.show()
```
